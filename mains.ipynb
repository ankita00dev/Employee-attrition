### Abstract
#### Emplyoee Attrition
Attrition is a process in which the workforce dwindles at a company, following a period in which a number of people retire or resign, and are not replaced.

A reduction in staff due to attrition is often called a hiring freeze and is seen as a less disruptive way to trim the workforce and reduce payroll than layoffs
Our aim is to analyze the dataset completely and find the reason behind Attrition of Employees and what are the top factors which lead to employee attrition?

### Description about the data
- left: The departure of employees from the organization.

- left: The departure of employees from the organization.

#### Machine Learning

```python
X = df.drop(['left'], axis=1)
y = df['left']
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state = 0)
X_over, y_over = rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_over, y_over, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

#### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()   
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

#### Model building

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
list_met = []
list_accuracy = []

#Model 1 - logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
# prediction
lr_pred_train = lr.predict(X_train)
lr_pred_test = lr.predict(X_test)
#accuracy
accuracy_LR = accuracy_score(Y_test, lr_pred_test)

#Model 2 - Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, Y_train) 
#prediction
dt_pred_train = dt.predict(X_train)
dt_pred_test = dt.predict(X_test)
# accuracy
accuracy_DT = accuracy_score(Y_test, dt_pred_test)

#Model 3 - RandomForest ginni
from sklearn.ensemble import RandomForestClassifier
rfg = RandomForestClassifier(random_state = 14)
rfg.fit(X_train, Y_train)
#predict
rfg_pred_train = rfg.predict(X_train)
rfg_pred_test = rfg.predict(X_test)
#evalutaion
accuracy_RFI = accuracy_score(Y_test, rfg_pred_test)

#Model 4 - random Forest entropy
rfe = RandomForestClassifier(criterion = 'entropy', random_state = 14)
rfe.fit(X_train, Y_train)
#predict
rfe_pred_train = rfe.predict(X_train)
rfe_pred_test = rfe.predict(X_test)
#evalutaion
accuracy_RFE = accuracy_score(Y_test, rfe_pred_test)

#Model 5 - SupportVectorMachine
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, Y_train)
#predict
svm_pred_train = svm.predict(X_train)
svm_pred_test = svm.predict(X_test)
#evalutaion
accuracy_SVC = accuracy_score(Y_test, svm_pred_test)

#Model 6 - KNN 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
#prediction
knn_pred_train = knn.predict(X_train)
knn_pred_test = knn.predict(X_test)
#evaluation
accuracy_knn = accuracy_score(Y_test, knn_pred_test)

list1 = ["Logistic Regression", "Decision Tree", "RandomForest Ginni", "RandomForest Entropy", 
         "Support Vector", "K Nearest Neighors",]

list2 = [accuracy_LR, accuracy_DT, accuracy_RFI, accuracy_RFE, accuracy_SVC, accuracy_knn]

df_accuracy = pd.DataFrame({'Method Used': list1, "Accuracy": list2})
print(df_accuracy)

charts = sns.barplot(x = 'Method Used', y = 'Accuracy', data = df_accuracy)
charts.set_xticklabels(charts.get_xticklabels(), rotation = 90)
```
